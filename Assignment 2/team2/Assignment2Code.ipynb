```python
import pandas as pd
dFrame = pd.read_csv('C:\Users\swethaj.2014\Desktop\hads2001.csv')
dFrame.head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
n=input('Please select the graph type: ')
if n==1:
    FinalData = dFrame[['Age','COSTMED']]
    completeData = FinalData.groupby(['Age']).COSTMED.sum() 
    my_plot=completeData.plot(kind='bar', title ="Cost by Age",figsize=(9,9),legend=True, fontsize=12)
elif n==2:
    FinalData = dFrame[['FMTREGION','FMTBEDRMS2','WEIGHT']]
    completeData = FinalData.groupby(['FMTREGION','FMTBEDRMS2']).WEIGHT.sum()
    my_plot=completeData.unstack().plot(kind='bar', stacked=True, title ="Weight by Region and Bedrooms",figsize=(9,9),legend=True, fontsize=12)
    plt.ylim(0,100000000)
elif n==3:
    FinalData = dFrame[['Age','TOTSAL']]
    completeData = FinalData.groupby(['Age']).TOTSAL.sum()
    my_plot=completeData.plot(title ="Salary by Age",figsize=(9,9),legend=True, fontsize=12)
```
