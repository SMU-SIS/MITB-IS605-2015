```python
import pandas as pd
dFrame = pd.read_csv('PythonData.csv')
```

```python
dFrame2015 = dFrame[(dFrame.Year==2015) & (dFrame.Products=='All')]
#2015
dFrame2015.head(10)
import pylab
import numpy as np
%matplotlib inline

pylab.xlabel('Month')
pylab.ylabel('Amount')
pylab.title('All product analysis for 2015')
pylab.plot(dFrame2015[['Month_number']],dFrame2015[['Gross_Sales']])
pylab.plot(dFrame2015[['Month_number']],dFrame2015[['Case_Deal']])
pylab.plot(dFrame2015[['Month_number']],dFrame2015[['Trade_Promotion _Other_RIR']])
pylab.plot(dFrame2015[['Month_number']],dFrame2015[['Over _Above']])
pylab.plot(dFrame2015[['Month_number']],dFrame2015[['Terms']])

#pylab.legend(['y = Gross_Sales', 'y = Case_Deal', 'y = Trade_Promotion', 'y = Over _Above','y = Terms'], loc='upper left')
#dFrame2015[['Year','Gross_Sales']]
```
