```python
import pandas as pd
from pandas import *
import pylab as pyl
from pylab import *
import csv
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame
import seaborn as sns
%matplotlib inline
import itertools
import Json

file = "Crimes_-_2001_to_present.csv"
for x in range(1,13):
    numRows = 100000*x
    df=pd.read_csv(file, nrows=numRows)
    print("I just loaded {} rows".format(numRows))
df.head()
```

```python
customers = df[['Primary Type','Year','FBI Code','District']]
customers.head()
category_group=customers.groupby(['Primary Type','Year']).sum()
category_group.head()
category_group.unstack().head()
my_plot = category_group.unstack().plot(kind='hist',stacked=True,title="Primary Type")
my_plot.set_xlabel("Primary Type")
my_plot.set_ylabel("Number")
```

```python
customers = df[['Location Description','Year','Arrest','District']]
customers.head()
category_group=customers.groupby(['District','Year']).sum()
category_group.head()
category_group.unstack().head()
my_plot = category_group.unstack().plot(kind='line',stacked=True,title=" The number of arrests in every District ")
my_plot.set_xlabel("District")
my_plot.set_ylabel("Numbers")
```
